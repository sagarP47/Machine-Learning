# Naive bayes

```python
from sklearn import naive_bayes
from sklearn import metrics
from sklearn import preprocessing
```

```python
X = data.drop('class',1)
y = data['class']
```

```python
X_norm = preprocessing.scale(X)
```

```python
nb = naive_bayes.GaussianNB()
```

```python
nb.fit(X_norm,y)
```

```python
nb_pred = nb.predict(X_norm)
```

```python
metrics.confusion_matrix(y,y_pred=nb_pred)
```

```python
print('Kappa Score is:',metrics.classification.cohen_kappa_score(y,nb_pred))
print('Accuracy is:',metrics.accuracy_score(y,y_pred=nb_pred))
```

```python
print(metrics.classification_report(y,nb_pred))
```

# Decision Tree

# Grid search

```python
from sklearn.model_selection import cross_val_score,GridSearchCV,train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_norm,y,stratify=y,test_size=0.3,random_state=123)
```

```python
grid_param = {'criterion':['gini','entropy'],'max_depth':[3,4,6]}
grid_tree = GridSearchCV(model_tree,param_grid=grid_param,cv=5,n_jobs=-1,scoring='accuracy')
grid_tree.fit(X_train,y_train)
cross_val_score(grid_tree,X_train,y_train,cv=5).mean()
```

```python
grid_tree.best_params_
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k_range = range(1,26)
weight_options = ['uniform','distance']
```

```python
grid_param_knn = {'n_neighbors': k_range,'weights': weight_options}
knn = KNeighborsClassifier()
```

```python
grid_knn = GridSearchCV(knn,param_grid=grid_param_knn,cv=5,n_jobs=-1,scoring='accuracy')
grid_knn.fit(X_train,y_train)
cross_val_score(grid_knn,X_train,y_train,cv=5).mean()
```

```python
grid_knn.best_params_
```

```python
knn1 = KNeighborsClassifier(n_neighbors=9,weights='distance')
knn1.fit(X_train,y_train)
knn_pred = knn1.predict(X_test)
```

```python
print(metrics.confusion_matrix(y_test,knn_pred))
print(metrics.cohen_kappa_score(y_test,knn_pred))
print(metrics.accuracy_score(y_test,knn_pred))
```

```python
print(metrics.classification_report(y_test,knn_pred))
```
